```python
#https://www.imdb.com/interfaces/

import pandas as pd
import numpy as np
title_basics = pd.read_csv(r"C:\Users\darkt\Dropbox\School\FA21\Codable\title.basics.tsv\title_basics.tsv", sep='\t')
title_crew = pd.read_csv(r"C:\Users\darkt\Dropbox\School\FA21\Codable\title.crew.tsv\title_crew.tsv", sep='\t')
title_principals = pd.read_csv(r"C:\Users\darkt\Dropbox\School\FA21\Codable\title.principals.tsv\title_principals.tsv", sep='\t')
title_ratings = pd.read_csv(r"C:\Users\darkt\Dropbox\School\FA21\Codable\title.ratings.tsv\title_ratings.tsv", sep='\t')
name_basics = pd.read_csv(r"C:\Users\darkt\Dropbox\School\FA21\Codable\name.basics.tsv\name_basics.tsv", sep='\t')
```

```python
title_basics_sub = title_basics.loc[(title_basics['titleType'] == 'movie') & (title_basics['runtimeMinutes'] != '\\N') ]
title_basics_sub = title_basics_sub.loc[(title_basics_sub['startYear'] != "\\N") & (title_basics_sub['endYear'] == "\\N")]
title_basics_sub['startYear'] = pd.to_numeric(title_basics_sub["startYear"])
title_basics_sub = title_basics_sub.loc[title_basics_sub['startYear'] >= 2000]
```

```python
title_basics_sub
```

```python
title_crew_sub = title_crew[title_crew.loc[:,'tconst'].isin(title_basics_sub.loc[:,'tconst'])]

title_crew_sub.loc[:,'directors'] = title_crew_sub.loc[:,'directors'].str.split(pat=",")
title_crew_sub = title_crew_sub.explode('directors')

title_crew_sub.loc[:,'writers'] = title_crew_sub.loc[:,'writers'].str.split(pat=",")
title_crew_sub = title_crew_sub.explode('writers')
```

```python
title_crew_sub
```

```python
title_principals_sub = title_principals[(title_principals['tconst'].isin(title_basics_sub['tconst'])) &
                 ((title_principals['nconst'].isin(title_crew_sub['directors'])) | 
                (title_principals['nconst'].isin(title_crew_sub['writers'])))]
```

```python
title_principals_sub = title_principals[title_principals['tconst'].isin(title_basics_sub['tconst'])]
```

```python
name_basics_sub = name_basics[
    name_basics['nconst'].isin(title_crew_sub['directors']) |
    name_basics['nconst'].isin(title_crew_sub['writers']) |
    name_basics['nconst'].isin(title_principals_sub['nconst'])
]

name_basics_sub
```
